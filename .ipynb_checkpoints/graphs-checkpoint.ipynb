# Helper Functions

```python
AMINO_ACIDS = 'ARNDCEQGHILKMFPSTWYV'

def index_array(idx: int, size: int):
    arr = np.zeros(size)
    arr[idx] = 1

    return arr

def ohe(iterable: Union[list, str], obs: Union[int, str]):

    return np.hstack([index_array(obs.index(i), len(obs)) for i in iterable])
```

```python
# ham distance and hd matrix functions

def ham_dist(q,r):
	return sum([1 for q,r in zip(q,r) if q != r])

def hd_matrix(qs: list):
	hd_i = []

	for q in qs:
		hd_i.append([ham_dist(q, r) for r in qs])

	return np.vstack(hd_i)
```

## Sequence entropy grid plot

```python
def position_entropy(aa_opts, base=None):
    """Return entropy of a list of optoins"""

    amino_acids, counts = np.unique(aa_opts, return_counts=True)
    return entropy(counts, base=base)

def sequence_entropy(msa: list[list]):
    """Return list of entropies for each position in an MSA."""

    msa = np.array(msa)

    return [position_entropy(msa[:,col]) for col in range(msa.shape[1])]

def seq_list(seqs: str):
    """Put list of str sequence into list of list format."""

    return [list(seq) for seq in seqs]

entropy_df = pd.DataFrame(columns=[0,1,2,3])

for model, num_mut in list(product(*[models, num_muts])):
    entropies = sequence_entropy(seq_list(gb1_vars[(gb1_vars.model == model) & (gb1_vars.num_muts == num_mut)].aa_sequence))
    position = list(range(len(entropies)))
    
    entropy_df = pd.concat([entropy_df, pd.DataFrame(np.array([position, entropies, [model]*len(entropies), [num_mut]*len(entropies)])).T], axis=0)

entropy_df.rename(columns={0:'position', 1:'entropy', 2:'model', 3:'num_muts'}, inplace=True)

```

```python
smuts = olson[olson.num_mutations == 1].copy()
smuts['position'] = [v[1:-1] for v in smuts.variant]
smuts.set_index('variant', inplace=True, drop=False)

def format_mutations(muts: str) -> list[str]:
    return "".join([m for m in muts if m not in "'()"]).split(',')
```

```python
def format_mutations(muts: str) -> list[str]:
    return "".join([m for m in muts if m not in "'() "]).split(',')

plots = ['lr', 'fcn', 'lcb', 'gcn']
model_names = {'lr':'LR',
               'fcn':'FCN',
               'lcb':'EnsC',
               'gcn':'GCN'}

fig, axs = plt.subplots(len(plots), 1, figsize=(15, 10), sharex=True, sharey=True)

for m, ax in zip(plots, axs.flatten()):
    unique_mutants = list(set(list(chain(*[format_mutations(mutants) for mutants in library[(library.model == m) & (library.num_muts == 10)].mutations]))))
    sns.scatterplot(data=smuts, x='position', y='score', ax=ax, linewidth=0, color='#5A5A5A')
    sns.scatterplot(data=smuts.loc[unique_mutants], x='position', y='score', ax=ax, color=color_dict[m], linewidth=1)
    ax.axhline(0, color='black')

    ax.set_title(f'Model: {model_names[m]}  |  proportion of mutations <= WT: {(len([s for s in smuts.loc[unique_mutants].score.tolist() if s <= 0]) / len(smuts.loc[unique_mutants])):.2f}')

# plt.savefig('single_mutations.png', dpi=300)
```
